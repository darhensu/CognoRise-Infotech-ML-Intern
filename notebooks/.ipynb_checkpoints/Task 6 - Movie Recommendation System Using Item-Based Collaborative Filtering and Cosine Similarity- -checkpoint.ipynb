# Building a Collaborative Filtering-Based Movie Recommendation System Using Cosine Similarity and TF-IDF

This project utilizes a content-based filtering approach combined with TF-IDF vectorization and cosine similarity to build a movie recommendation system. By processing and analyzing metadata such as genres, keywords, cast, and crew, the system identifies movies with similar features and suggests them to users based on their preferences. With the help of the cosine similarity metric, the model finds movies that share thematic or stylistic elements with the input movie. This recommendation system enhances user experience by providing personalized movie suggestions, enabling users to discover new films aligned with their interests.

# Import Libraries

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

# Load and Explore Dataset

```python
df1 = pd.read_csv(r'C:\Users\user\Desktop\CognoRise Infotech ML Intern\data\tmdb_5000_movies_data.csv')
df1.head()
```

```python
df2 = pd.read_csv(r'C:\Users\user\Desktop\CognoRise Infotech ML Intern\data\tmdb_5000_credits_data.csv')
df2.head()
```

# Merge/Join Tables

```python
df = pd.merge(df1,df2, left_on='id', right_on='movie_id')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Data Preprocessing

```python
# fill missing values in relevant columns
df['overview'] = df['overview'].fillna('')
df['runtime'] = df['runtime'].fillna(df['runtime'].mean())  # replace NaNs with mean runtime
df['genres'] = df['genres'].fillna('')
df['keywords'] = df['keywords'].fillna('')
df['cast'] = df['cast'].fillna('')
df['crew'] = df['crew'].fillna('')
```

# Feature Engineering

```python
# combine selected columns to create a single string of features for each movie
df['combined_features'] = df['genres'] + " " + df['keywords'] + " " + df['cast'] + " " + df['crew'] + " " + df['overview']
```

```python
# convert the combined text to a matrix of TF-IDF features
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['combined_features'])
```

```python
# generate the TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(df['combined_features'])
```

```python
# calculate cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

# Recommendation Function

```python
# function to get movie recommendations based on title
def get_recommendations(title, cosine_sim=cosine_sim):
    # find the index of the movie with the given title
    idx = df[df['title_x'] == title].index[0]
    
    # get the pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # sort movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # get the indices of the 10 most similar movies (excluding the first one)
    sim_indices = [i[0] for i in sim_scores[1:11]]
    
    # return the top 10 most similar movies
    return df['title_x'].iloc[sim_indices]
```

# Results

```python
movie_title = 'The Dark Knight'  # replace with a movie title to what u want in the dataset
recommended_movies = get_recommendations(movie_title)
print("Movies recommended for:", movie_title)
print(recommended_movies)
```
